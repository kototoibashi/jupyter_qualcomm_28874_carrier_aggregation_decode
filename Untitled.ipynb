```python
# decoding 28874
# ref: https://github.com/HandyMenny/28874Decoder

import mmap

```

```python
def decode_desc201(mm,fptr):
    descok = 0
    for i in range(6):
        item = int.from_bytes(mm[fptr:fptr+2],'little')
        fptr += 2
        print('BAND: ',item)
        if item != 0:
            descok = 1
        item = int.from_bytes(mm[fptr:fptr+1],'little')
        fptr += 1
        print('Bclass: ',item)
        item = int.from_bytes(mm[fptr:fptr+1],'little')
        fptr += 1
        print('Ant: ',item)
    return fptr

def decode_desc202(mm,fptr):
    descok = 0
    item = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('BAND: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('Bclass: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('buf: ',item)
    item = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('BAND: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('Bclass: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('buf: ',item)
    item = mm[fptr:fptr+16]
    fptr += 16
    print('buf: ',item)
    return fptr

def decode_desc(mm,fptr):
    print('---')
    print('start descriptor')
    desctype = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('type: ',desctype)
    if desctype == 201:
        fptr = decode_desc201(mm,fptr)
    elif desctype == 202:
        fptr = decode_desc202(mm,fptr)
    return fptr

with open("unpacked", "rb") as f:
    mm = mmap.mmap(f.fileno(), 0, access=mmap.ACCESS_READ)
    fptr = 0
    print('Format version',int.from_bytes(mm[0:2],'little'))
    print('Number of descriptors: ',int.from_bytes(mm[2:4],'little'))
    fptr = 4
    for i in range(142):
        fptr = decode_desc(mm,fptr)
        print('fptr: ',fptr)


    
```
