```python
def decode_desc201(mm,fptr):
    print("DOWNLINK COMBO")
    for i in range(6):
        item = int.from_bytes(mm[fptr:fptr+2],'little')
        fptr += 2
        print('BAND: ',item)
        if item != 0:
            descok = 1
        item = int.from_bytes(mm[fptr:fptr+1],'little')
        fptr += 1
        print('Band class: ',item)
        item = int.from_bytes(mm[fptr:fptr+1],'little')
        fptr += 1
        print('Antenna: ',item)
    return fptr

def decode_desc202(mm,fptr):
    print("UPLINK COMBO")
    item = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('BAND: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('Band class: ',item)
    item = mm[fptr:fptr+1]
    fptr += 1
    print('(padding ',item,')')
    item = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('BAND: ',item)
    item = int.from_bytes(mm[fptr:fptr+1],'little')
    fptr += 1
    print('Bclass: ',item)
    item = mm[fptr:fptr+1]
    fptr += 1
    print('(padding ',item,')')
    item = mm[fptr:fptr+16]
    fptr += 16
    print('(padding ',item,')')
    return fptr

def decode_desc(mm,fptr):
    print('---')
    desctype = int.from_bytes(mm[fptr:fptr+2],'little')
    fptr += 2
    print('descriptor type: ',desctype)
    if desctype == 201:
        fptr = decode_desc201(mm,fptr)
    elif desctype == 202:
        fptr = decode_desc202(mm,fptr)
    return fptr

with open("unpacked", "rb") as f:
    mm = mmap.mmap(f.fileno(), 0, access=mmap.ACCESS_READ)
    fptr = 0
    print('Format version',int.from_bytes(mm[0:2],'little'))
    dcount = int.from_bytes(mm[2:4],'little')
    print('Number of descriptors: ',dcount)
    fptr = 4
    for i in range(dcount):
        fptr = decode_desc(mm,fptr)
        #print('fptr: ',fptr)


    
```
